```python
#Removing square brackets
def concatenate_choices(choices):
    if pd.notna(choices):
        # Remove square brackets and quotes, if present
        choices = choices.replace('[', '').replace(']', '').replace("'", "")
        # Split the choices and join them with a comma
        choices_list = choices.split(', ')
        return ', '.join(choices_list)
    else:
        return None

# Apply the function to the 'choice_description' column
df['choice_description'] = df['choice_description'].apply(concatenate_choices)

# Display the modified DataFrame
print(df[['order_id', 'item_name', 'choice_description']])
```

```python
df.head(5)
```

```python
# Function to extract individual choices
def extract_choices(choices, choice_number):
    if pd.notna(choices):
        # Remove square brackets and quotes, if present
       # choices = choices.replace('[', '').replace(']', '').replace("'", "")
        # Split the choices and return the choice at the specified index
        choices_list = choices.split(', ')
        return choices_list[choice_number] if choice_number < len(choices_list) else None
    else:
        return None

# Define the maximum number of choices you want to extract
max_choices = 5

# Create new columns for each choice
for i in range(max_choices):
    df[f'choice_{i+1}'] = df['choice_description'].apply(lambda x: extract_choices(x, i))

# Display the modified DataFrame
print(df[['order_id', 'item_name', 'choice_description', 'choice_1', 'choice_2', 'choice_3', 'choice_4', 'choice_5']])
```
